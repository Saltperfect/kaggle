```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import re
import os
```

```python
def get_data(station,year,month,date):
    url = 'http://weather.uwyo.edu/cgi-bin/sounding?region=naconf&TYPE=TEXT%3ALIST&YEAR='+str(year)+'&MONTH='+str(month)+'&FROM='+str(date)+'00&TO='+str(date)+'00&STNM='+str(station)
    path = 'weather/'+str(station)+'_'+str(year)+'_'+str(month)+'_'+str(date)+'.txt'
    if(os.path.isfile(path)):
        print('file exists',path)
        return 
    while(True):
        try:
            urlData = requests.get(url).content
    #     soup = BeautifulSoup(urlData,"lxml")

    #         table = soup.find("pre").find(text=True)
            
            if("Invalid TIME parameter." not in urlData):
                f = open(path,'w+')
                f.write(urlData)
                f.close()
            break
        except Exception as  e:
            print e
```

```python
stations = ['16080', '07145', '10739', '10868', '06610']
# get_data(,stations[0],2015,2,31)
```

```python
for station in stations:
    for year in range(2010,2018):
        for month in range(1,13):
            for date in range(1,32):
                get_data(station,year,month,date)
```
